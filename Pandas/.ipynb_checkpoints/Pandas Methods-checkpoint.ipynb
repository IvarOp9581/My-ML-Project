# ***Apply Method:***
***`Apply` Method on Single Column:***

```python
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(123456789)
```

```python
df['CC last_four']=df['CC Number'].apply(last_four)
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

```python
df.head()
```

***`Apply` Method on Multiple Column:***

```python
# Lambda Expression
lambda num: num*2
```

```python
df['total_bill'].apply(lambda bill: bill*2)
```

```python
df.head()
```

```python
def quality_of_tip(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
quality_of_tip(16.99, 1.01)
```

```python
df["Quality"]= df[['total_bill', 'tip']].apply(lambda df: quality_of_tip(df['total_bill'], df['tip']), axis=1)
```

```python
# How to run the above code faster and simpler
df['Quality']=np.vectorize(quality_of_tip)(df['total_bill'], df['tip'])
```

```python
df.head()
```

```python
# Let's prove the vectorize function runs faster using `timeit`
import timeit
```

```python
# code snippet to be executed only once 
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
'''
```

```python
# code snippet whose execution time is to be measured 
stmt_one = ''' 
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
'''

stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
```

```python
timeit.timeit(setup = setup, stmt = stmt_one, number = 1000)
```

```python
timeit.timeit(setup = setup, stmt = stmt_two, number = 1000)
```
